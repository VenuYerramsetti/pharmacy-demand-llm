## 📥 Load all datasets

We load:
- `sales_daily.csv`
- `sales_weekly.csv`
- `sales_monthly.csv`
- `sales_hourly.csv`

These files live in the `/data` folder at the project root.

```python
# ===============================
# Load all datasets from /data
# ===============================

base = "../data/"

df_daily   = pd.read_csv(base + "sales_daily.csv")
df_weekly  = pd.read_csv(base + "sales_weekly.csv")
df_monthly = pd.read_csv(base + "sales_monthly.csv")
df_hourly  = pd.read_csv(base + "sales_hourly.csv")

df_daily.head(), df_weekly.head()
```

## 🛠 Step 1 — Convert date columns

```python
# Convert date columns in each dataset
df_daily['datum'] = pd.to_datetime(df_daily['datum'])
df_weekly['datum'] = pd.to_datetime(df_weekly['datum'])
df_monthly['datum'] = pd.to_datetime(df_monthly['datum'])
df_hourly['datum'] = pd.to_datetime(df_hourly['datum'])

print("Date columns successfully converted.")
```

## 🧹 Step 2 — Check for missing values

```python
print("Daily missing values:\n", df_daily.isna().sum())
print("Weekly missing values:\n", df_weekly.isna().sum())
print("Monthly missing values:\n", df_monthly.isna().sum())
print("Hourly missing values:\n", df_hourly.isna().sum())
```

## 📊 Step 3 — Basic dataset info

```python
df_daily.info()
df_weekly.info()
df_monthly.info()
df_hourly.info()
```

## 🧩 Step 4 — Create basic features

```python
# Function to apply to all
    
def add_date_features(df):
    df['Year'] = df['datum'].dt.year
    df['Month'] = df['datum'].dt.month
    df['Day'] = df['datum'].dt.day
    df['Weekday'] = df['datum'].dt.weekday
    df['Weekday_Name'] = df['datum'].dt.day_name()
    return df

df_daily = add_date_features(df_daily)
df_weekly = add_date_features(df_weekly)
df_monthly = add_date_features(df_monthly)
df_hourly = add_date_features(df_hourly)

df_daily.head()
```

## 💾 Step 5 — Save cleaned datasets

```python
os.makedirs("outputs/cleaned", exist_ok=True)
```

```python
df_daily.to_csv("outputs/cleaned/daily_cleaned.csv", index=False)
df_weekly.to_csv("outputs/cleaned/weekly_cleaned.csv", index=False)
df_monthly.to_csv("outputs/cleaned/monthly_cleaned.csv", index=False)
df_hourly.to_csv("outputs/cleaned/hourly_cleaned.csv", index=False)

print("All cleaned datasets saved successfully.")
```
