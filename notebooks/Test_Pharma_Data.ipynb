````python
df_daily = pd.read_csv('../data/sales_daily.csv')
df_weekly = pd.read_csv('../data/sales_weekly.csv')
df_monthly = pd.read_csv('../data/sales_monthly.csv')
df_hourly = pd.read_csv('../data/sales_hourly.csv')

df_daily.head()
````

````python
df = df_daily.copy()

# Convert date column
df['datum'] = pd.to_datetime(df['datum'], dayfirst=False)

df.info()
df.head()
````

````python
df.isna().sum()
````

````python
df.describe()
````

````python
import matplotlib.pyplot as plt

category_cols = ['M01AB','M01AE','N02BA','N02BE','N05B','N05C','R03','R06']

df[category_cols].sum().sort_values(ascending=False).plot(kind='bar', figsize=(10,5))
plt.title("Total Sales by Medication Category")
plt.ylabel("Total Units Sold")
plt.show()
````

````python
weekday_sales = df.groupby("Weekday Name")[category_cols].sum()
weekday_sales.sum(axis=1).plot(kind='bar', figsize=(10,5))
plt.title("Total Sales by Weekday")
plt.ylabel("Units Sold")
plt.show()
````

# Plot Hourly Sales Trend

````python
df.groupby("Hour")[category_cols].sum().sum(axis=1).plot(figsize=(12,5))
plt.title("Hourly Sales Trend")
plt.ylabel("Units Sold")
plt.show()
````
